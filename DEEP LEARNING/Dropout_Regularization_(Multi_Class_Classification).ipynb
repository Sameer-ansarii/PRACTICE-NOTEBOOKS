# Dropout Regularization in Deep Neural Network

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv(r'Iris.csv')
df.head()
```

```python
df.shape
```

```python
df.Species.value_counts()
```

```python
df['Species'] = df['Species'].replace(['Iris-setosa', 'Iris-versicolor', 'Iris-virginica'], [0,1,2])
```

```python
X = df.drop(['Id','Species'],axis=1)
y = df['Species']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
X_train.head()
```

```python
X_test.head()
```

```python

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

# Using Deep Learning Model

### Model without using Dropout

```python
import tensorflow as tf
from tensorflow import keras
from keras.models import Sequential
from keras.layers import Dense, Dropout
```

```python
model = Sequential([
    Dense(68, input_dim=4, activation='relu'),
    Dense(1, activation='softmax')
])

model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

model.fit(X_train_scaled, y_train, epochs=100)
```

```python
model.evaluate(X_test, y_test)
```

**Training Accuracy >>> Test Accuracy**

```python
y_pred = model.predict(X_test).reshape(-1)
print("Predicted Probabilities:")
print(y_pred[:10])

y_pred_rounded = np.round(y_pred)
print("Rounded Predictions (0 or 1):")
print(y_pred_rounded[:10])
```

```python
y_test[:10]
```

```python
from sklearn.metrics import confusion_matrix, classification_report

# Calculate the confusion matrix
cm = confusion_matrix(y_test, y_pred)

# Print the classification report
print("Classification Report:")
print(classification_report(y_test, y_pred))

# Print the confusion matrix
print("Confusion Matrix:")
print(cm)
```

# Model with using Dropout

```python
model = Sequential([
    Dense(60, input_dim=4, activation='relu'),
    Dropout(0.3),
    Dense(30, activation='relu'),
    Dropout(0.3),
    Dense(15, activation='relu'),
    Dropout(0.3),
    Dense(1, activation='softmax')
])

model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

model.fit(X_train_scaled, y_train, epochs=100)
```

```python
model.evaluate(X_test, y_test)
```

```python
y_pred = model.predict(X_test).reshape(-1)
print("Predicted Probabilities:")
print(y_pred[:10])

y_pred_rounded = np.round(y_pred)
print("Rounded Predictions (0 or 1):")
print(y_pred_rounded[:10])
```

```python
y_test[:10]
```

```python
# Calculate the confusion matrix
cm = confusion_matrix(y_test, y_pred)

# Print the classification report
print("Classification Report:")
print(classification_report(y_test, y_pred))

# Print the confusion matrix
print("Confusion Matrix:")
print(cm)
```
