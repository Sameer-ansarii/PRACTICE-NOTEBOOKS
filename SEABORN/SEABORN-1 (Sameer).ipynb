# Histogram Plot

```python
# How to create histogram plot

sns.displot(x["flipper_length_mm"])
plt.show()
```

```python
# Histogram plot creates unorganised X axis to resolve them we use bins=(x axis index)
# Kernal Density Estimator:- kde use for creating line inside the histogram which indicates estimate value of data. kde=True
# rug=True parameter use for showing the scale on x axis. Useful with kde parameter
# To change the color of Histogram Plot we use color = "color shortcut" parameter

sns.displot(x["flipper_length_mm"],bins=[175,180,185,190,195,200],kde=True,rug=True,color="k")
plt.show()
```

```python
# To create the histogram with log scale we use log_scale=True

sns.displot(x["flipper_length_mm"],log_scale=True)
plt.show()
```

# Bar Plot 

```python
# We import full data set for understanding things better
# For creating grid in backgroud of bar plot we use plt.grid() function

x = sns.load_dataset("penguins")

sns.barplot(x="island",y="bill_depth_mm",data=x)
plt.grid()
plt.show()
```

```python
# To divide bar plot between the 3rd column. We use hue="3rd column name"
# It is use at the time of data visualization when we divide the particular data between 2 parameters 
# To change the order and name on X axis we use order= variable in which we set the data order according to x axis data 
# We change change the style of grid by using sns.set(style="dark,darkgrid") function

order1 = ["Dream","Torgersen","Biscoe"]
sns.barplot(x="island",y="bill_depth_mm",data=x,hue="sex",order=order1)

plt.show()
```

```python
# To change the position of data of hue parameter we use hue_order=["category1","catogery2"....]
# You notice that there is a line above all the bars. To change the size of that line we use ci=(0-100). The default size is 100

sns.barplot(x="island",y="bill_depth_mm",data=x,hue="sex",hue_order=["Female","Male"],ci=0)
plt.show()
```

```python
# How to create horizontal histogram plot. We use orient="h"
# NOTE:- The value of X axis and Y axis should be in numerical form otherwise it will show error

sns.barplot(x="island",y="bill_depth_mm",data=x,hue="sex",orient="h")
plt.show()
```

```python
# This bar plot shown same in both shape vertical and horizontal because it contains only numerical data 

sns.barplot(x="flipper_length_mm",y="bill_depth_mm",data=x,hue="sex",orient="h")
plt.show()
```

```python
# To change the color of bars we use color = "color shortcut".
#  If we use the color parameter all the bars turns into the same color
# To adjust the saturation of bars we use saturation=__

sns.barplot(x="island",y="bill_depth_mm",data=x,color="r",saturation=0.4)
plt.show()
```

```python
# To change the color according color bar we use palette="choose from color bar"
# To change the color of line above the bars we use errcolor="color shortcut"
# To change the width of this line we use errwidhth=__

sns.barplot(x="island",y="bill_depth_mm",data=x,palette="Accent",hue="sex",errcolor="r",errwidth=10)
plt.show()
```

```python
# To create the cap above and below this line we use capsize=__. By default the cap size is 0 thats why the cap is not showing

sns.barplot(x="island",y="bill_depth_mm",data=x,palette="Accent",capsize=0.1)
plt.show()
```

```python
# To overlap the data of catogeries in the hue parameter we use dodge=False
# To adjust the brightness of bar plot we use aplha=(0-1). By default it is 1

sns.barplot(x="island",y="bill_depth_mm",data=x,hue="sex",dodge=False,alpha=0.7)
plt.show()
```

# Scatter Plot

```python
# How to create scatter plot. We can create scatter plot with one numerical and one object column also.
# But generally we use scatter plot to compare two numerical data

x = sns.load_dataset("penguins")
sns.scatterplot(x="island",y="bill_depth_mm",data=x)
plt.show()
```

```python
# To change the sizes of dots which indicates different catogries 
# we use hue="column name",style="column name",size="column name", sizes=(20,40...)
# Each parameter is cumplusory for customize otherwise size will not change
# To change the color of dots we use palette="choose color from color bar"

sns.scatterplot(x="bill_depth_mm",y="bill_length_mm",data=x,hue="sex",style="sex",size="sex",sizes=(150,20),palette="flare")
plt.show()
```

# Heat Map

## Heat Map waorks on 2D Data not on 1D Data. Heap Map mostly used in Machine Learning

```python
# Lets create 2D array

x = np.linspace(0,20,20).reshape(5,4)
x
```

```python
# How to create heatmap

sns.heatmap(x)
plt.show()
```

```python
# How to change the starting and ending value of colorbar. We use vmin=starting point, xmax=ending point
# How to change the color of heatmap. We use cmap=choose color from colorbar
# To show the values of data inside the heatmap we use annot=True

sns.heatmap(x,vmin=0,vmax=26,cmap="Blues",annot=True)
plt.show()
```

```python
# How to change the values showing inside the heatmap. 
# Lets create a 2D because heatmap works on 2D Data. We use annot=variable of changed data, fmt="data type of new data"
# Remember that new data is equivalent in number to old data otherwise it will show error
# How to change the color and fontsize of data inside the heatmap. We use annot_kws=variable in which we customize data
# How to increase the width of boxes inside the heatmap to seprate them. We use linewidth=__.By default it is 0

v = {"fontsize":20,"color":"g"}
var = np.array([["a1","a2","a3","a4"],["b1","b2","b3","b4"],["c1","c2","c3","c4"],["d1","d2","d3","d4"],["e1","e2","e3","e4"]])

sns.heatmap(x,vmin=0,vmax=26,cmap="Blues",annot=var,fmt="s",annot_kws=v,linewidth=1)
plt.show()
```

```python
# How to get a colored outline over the boxes. We use linecolor="color shortcut"
# How to remove color bar from heatmap. We use cbar=False
# To remove labels of X and Y axis we use xticklabels=False, yticklabels=False
# To change the name of x and y axis label we use .set(xlabel="__",ylabel="__")
# To change the fontsize we use sns.set(font_scale=__)


v = sns.heatmap(x,cmap="Blues",annot=var,fmt="s",linewidth=1,linecolor="r",cbar=False,xticklabels=False,yticklabels=False)
v.set(xlabel="Python",ylabel="C++")
sns.set(font_scale=1)
plt.show()
```
