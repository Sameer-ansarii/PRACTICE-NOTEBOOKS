# 4) Distplot (Numerical-Categorical)

It is used to see the distribution of data. 

By default it creates Probability Density Function (PDF) inside the Distplot

```python
sns.distplot(titanic['Age'])
plt.show()
```

We see here the distribution of people by age who survived

```python
# There is no hue parameter in distplot so we create it by self by using this code

sns.distplot(titanic[titanic['Survived']==0]['Age'])
```

We see here the distribution of people by age who are not survived

```python
# We can overlap the different categorices of variable inside one distplot

sns.distplot(titanic[titanic['Survived']==0]['Age'])
sns.distplot(titanic[titanic['Survived']==1]['Age'])
plt.show()
```

We see here the distribution of people by age and gender who are survived or not

```python
# To avoid the histogram create inside distplot to see the bell curve better and understand distribution better

sns.distplot(titanic[titanic['Survived']==0]['Age'],hist=False)
sns.distplot(titanic[titanic['Survived']==1]['Age'],hist=False)
plt.show()
```

We see here the probability of surviving by age and sex

# 5) Heatmap (Categorical-Categorical)

Question - How many people died and survived in each class ticket

```python
pd.crosstab(titanic['Pclass'],titanic['Survived'])
```

In 1st class 80 people died and 136 people survived.


Conclusion :- Their are more people in 1st class who survived but in 2nd and 3rd class more people died and less people survived

```python
sns.heatmap(pd.crosstab(titanic['Pclass'],titanic['Survived']),annot=True)
```

light color in the heatmap indicates the people who died and dark color indicates the number of people who survived

Conclusion :- We see this there are more number of people who died in 3rd class as compare to 1st and 2nd class

```python
titanic.groupby('Pclass').mean()['Survived']*100
```

We see here the percentage of people who survived in each class


Conclusion :- There is higher percentage of people in 1st class who survived

```python
titanic.groupby('Sex').mean()['Survived']*100
```

We see here the population of female is more than male population who survived 

```python
titanic.groupby('Embarked').mean()['Survived']*100
```

The more percentage of people who survived starts their journey from Camberk

# 7) Lineplot (Numerical-Numerical)

We use lineplot if their is a time series on X-axis like month,date,year. Eg :- Covid Numbers, Stock Inflation Chart, etc.

```python
flights.head()
```

Question :- Number of passengers travelled every year.

```python
flights.groupby('year').sum().reset_index()
```

We see that in 1949 the number of passengers travelled is 1520


```python
x = flights.groupby('year').sum().reset_index()
sns.lineplot(x['year'],x['passengers'])
```

We see here that number of passengers increase year by year and their is a linear growth of airline company

```python
# Use of pivot_table in EDA

flights.pivot_table(values='passengers',index="month",columns="year")
```

```python
sns.heatmap(flights.pivot_table(values='passengers',index="month",columns="year"))
```

We see here that in july 1960 more passengers travelled. It is a highest revenue generated month of all time

More people travell during September to June

```python
sns.clustermap(flights.pivot_table(values='passengers',index="month",columns="year"))
```

Clustermap matches the columns who behave similarly. We easily analyize the relationship between columns
