```python
# Will continue as long as the condition is true
i=1
while i<5:
    print(i)
    i=i+1
print('while loop has ended')    
```

```python
#Infinite Loop
i=1
while True:
    print(i,end='')
    i=i+1
```

```python
# Break (Breaks Out of loop)
i=1
while True:
    if i==10:
        break # It continues as long as if condition is False it stops when the if condition is True
    print(i)
    i=i+1
print("Hello")
```

```python
#Continue (Takes back to loop)
i=1
while i<=10:
    if i==6:
        i=i+1 # We have to tell here what to do and continue
        continue
        print("A")
        print("b")
    else:
        print(i,end=' ')
        i=i+1        
```

```python
s="Arihant Jain"
for i in s:
    if i == 't':
        continue
    print(i)
else:
    if i=="t":
        print("I got t")
print("Good job")
```

# Functions

* Functions are useful for breaking up a large program to make it easier to read and maintain. 


* They are also useful to avoid writing the same code at several different points in your program.


* Function is a block of code that performs a specific task. 


* This helps in avoiding redundancy in code

Lambda function
A lambda function is a small anonymous function.
A lambda function can take any number of arguments, but can only have one expression.

lambda arguments : expression


```python
Example
Add 10 to argument a, and return the result:
Lambda functions can take any number of arguments:
```

```python
x = lambda a : a + 10
print(x(5))
```

```python
Example
Multiply argument a with argument b and return the result:
```

```python
x = lambda a, b : a * b
print(x(5, 6))
```
