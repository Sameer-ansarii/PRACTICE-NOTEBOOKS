# Count Plot

## What's the difference between Count Plot and Bar Plot?
## Countplot plots the count of the number of records by category.
## Barplot plots a value or metric for each category (by default, bar plots the mean of a             variable, by category). 

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd 
```

```python
var = sns.load_dataset("tips")
var.head()
```

```python
# How to make a Count Plot. We use sns.countplot(x="Column name",data=Variable of data frame)

sns.countplot(x="sex",data=var)
plt.show()
```

```python
# How to use hue parameter
# To change the color of count plot we use color="color shortcut"
# To adjust the color level of bars we use saturation=0-1. By default it is 1

sns.countplot(x="sex",data=var,hue="smoker",color="r",saturation=0.3)
plt.show()
```

```python
# How to make horizontal count plot
# To change the color of bars we use palette="choose from color bar" 

sns.countplot(y="sex",data=var,hue="smoker",palette="Blues")
plt.show()
```

# Violin Plot 

## A Violin Plot plays a similar role as a Box and Whisker Plot. It shows the distribution of quantitative data across several levels of one (or more) categorical variables such that those distributions can be compared.

```python
# How to create a violin plot
# X axis = object data type column, y axis = numerical data type column

sns.violinplot(x="day",y="total_bill",data=var)
plt.show()
```

```python
# How to use hue="__" parameter in violin plot

sns.violinplot(x="day",y="total_bill",data=var,hue="time")
plt.show()
```

```python
# To increase the width of outlines in violin plot we use linewidth=__
# To change the color of violin boxes we use palette="choose from color bar"

sns.violinplot(x="day",y="total_bill",data=var,hue="time",linewidth=3,palette="BuGn")
plt.show()
```

```python
# How to create violin plot according to the categories of particular column we use
#(x="column name from which you categorised data", y="column 2 name use for giving situation,data=var,order=["cat1",cat2,..])
# To adjust the color of violin boxes we use saturation=___
# To change the color of all boxes as same we use color="color shortcur"

sns.violinplot(x="time",y="total_bill",data=var,order=["Dinner","Lunch"],saturation=0.4,color="r")
plt.show()
```

```python
# To split data we use split=True
# To change the shape of of box inside the violin box we use inner=("stick","quartile","point")

sns.violinplot(x="day",y="total_bill",data=var,hue="sex",split=True,inner="stick")
plt.show()
```

```python
# To create horizontal violin plot x axis = numerical column y axis = object column

sns.violinplot(x="total_bill",y="day",data=var)
plt.show()
```

```python
# How to create a single violin plot according to data of a particular column
# sns.violinplot(y=var of DF["Numerical column Name"])

sns.violinplot(y=var["total_bill"])
plt.show()
```

# Pair Plot

```python
# How to create Pair Plot
# We use less data to understand things clear

var = sns.load_dataset("tips").head(10)

sns.pairplot(var)
plt.show()
```

```python
# How to use hue="column name" parameter in pair plot

sns.pairplot(var,hue="sex")
plt.show()
```

```python
# How to create pair plot of particular column
# We use vars=["column name 1","column name 2"...]
# To change the shape of dots inside the scatter plot of pair plot we use markers=[">",">"...]

sns.pairplot(var,vars=["total_bill","tip"],hue="sex",markers=[">","*"])
plt.show()
```

```python
# How to change the order of hue parameter. We use hue_order=["catogry1","category2"..]
# To change the color of pair plor using palette parameter we use palette="choose from color bar"

sns.pairplot(var,hue="sex",hue_order=["Female","Male"],vars=["total_bill","tip"],palette="BuGn")
plt.show()
```

```python
# How to customize the data on X axis, Y axis. We use x_vars=["column name1","column name 2"..]

sns.pairplot(var,x_vars=["total_bill","tip"],y_vars=["day","size"])
plt.show()
```

```python
# To change the type of pair plot we use kind=("scatter","kde","hist","reg")

sns.pairplot(var,kind="kde") # "scatter","kde","hist","reg".
plt.show()
```

```python
# To change the type of pair plot we use kind=("scatter","kde","hist","reg")
# You notice that diagonal plots in pair plot are of different kind. To change their kind we use diag_kind=("hist","kde")

sns.pairplot(var,kind="kde",diag_kind="hist") # "kde","hist"
plt.show()
```

# Strip Plot

## A Strip Plot is very simple to understand. It is basically a scatter plot that differentiates different categories. Strip Plots are considered a good alternative to box or a violin plot. 

```python
# How to create a Strip plot

sns.stripplot(x="day",y="total_bill",data=var)
plt.show()
```

```python
# How to create a pair plot of single column

sns.stripplot(y=var["total_bill"])
plt.show()
```

```python
# How to use hue parameter in strip plot. We use hue="column name"
# To adjust the brightness of pairplot we use alpha=(0-1). By Default it is 1

sns.stripplot(x="day",y="total_bill",data=var,hue="sex",alpha=0.5)
plt.show()
```

```python
# To change the color of strip plot we use palette="choose from colorbar" parameter
# To increase the outline of dots we use linewidth=__
# To change the color of outline of dots we use edgecolor="color shortcut"

sns.stripplot(x="day",y="total_bill",data=var,hue="sex",palette="Blues",linewidth=1,edgecolor="r")
plt.show()
```

```python
# How to split the distribution of dots. We use jitter=__
# To change the size of dots we use size=__
# How to change the shape of dots we use marker="*",">","<"

sns.stripplot(x="day",y="total_bill",data=var,hue="sex",palette="Blues",jitter=2,size=10,marker="*")
plt.show()
```

# Box Plot

```python
var = sns.load_dataset("tips")
var.head()
```

```python
# How to create a box plot.
# sns.boxplot(x="object column",y="numerical column",data=variable of data set)

sns.boxplot(x="day",y="total_bill",data=var)
plt.show()
```

```python
# How to create grids inside box plot

sns.set(style="whitegrid")
sns.boxplot(x="day",y="total_bill",data=var)
plt.show()
```

```python
# How to use hue parameter in box plot. hue="columnn name"
# To change the color of boxes we use color="color shortcut"
# To change the order of X axis we use order=["cat. 1.", "cat. 2"....]
# To create a symbol at the place of mean inside boxes we use showmeans=True 

sns.boxplot(x="day",y="total_bill",data=var,hue="sex",color="g",order=["Fri","Sun","Thur","Sat"],showmeans=True)
plt.show()
```

```python
# To change the properties of symbol we use meanprops={"marker":"*","markeredgecolor":"m"}
# "marker"="_" use for change the shape of symbol. "markeredgecolor":"color shortcut" use for outline the symbol
# linewidth=_ is use for adjust the width lines of boxes

sns.boxplot(x="day",y="total_bill",data=var,showmeans=True,meanprops={"marker":"*","markeredgecolor":"r"},color="w",linewidth=3)
plt.show()
```

```python
# To change the color of box plot using color from color bar. We use palette="choose color from color bar"

sns.boxplot(x="day",y="total_bill",data=var,palette="BuGn")
plt.show()
```

```python
# To make horizonatl box plot we use sns.boxplot(x="numerical column"y="object column")

sns.boxplot(x="total_bill",y="day",data=var)
plt.show()
```

```python
# How to create a box of particular column
# sns.boxplot(y=var of DF["column name"])

sns.boxplot(y=var["total_bill"])
sns.set(style="white")
plt.show()
```

```python
# How to use hue="column name" parameter in catplot
# To change the color of dots we use palette="choose from color bar" parameter

sns.catplot(x="tip",y="size",data=var,hue="sex",palette="BuGn")
plt.show()
```

```python
# To change the height of catplot we use height=_ parameter. 
# You see on x axis, values of x axis are overlaping this height parameter is used to sort this

sns.catplot(x="tip",y="size",data=var,hue="sex",palette="BuGn",height=10)
plt.show()
```

# Styling In Plots

## Seaborn Figure Styles
## Removing axes Spines
## Scale and Context

```python
# How to create simple grid inside plots. We use plt.grid() function 

var = sns.load_dataset("tips")

sns.barplot(x="day",y="total_bill",data=var)
plt.grid()
plt.show()
```

```python
# How to create grid in background of different styles using seaborn
# sns.set_style("white","whitegrid","dark","darkgrid","ticks")
# We also use matplotlib grid function and seaborn grid function together

sns.set_style("darkgrid")
sns.barplot(x="day",y="total_bill",data=var)
plt.grid()
plt.show()
```

```python
# To remove the opposite line of x and y axis we use sns.despine() function. We use this to see plot better 
# To change the size of plot we use plt.figure(figsize=(width,height)) function

sns.set_style("white")
plt.figure(figsize=(10,5))
sns.barplot(x="day",y="total_bill",data=var)
sns.despine()
plt.show()
```

```python
# For outstyling of plot we use sns.set_contect("paper","notebook","talk","poster",font_scale=_) we font_scale=_ to increase/decrease size of text
# By default font_scale is 1
# To change the color of plot according color bar we use palette="choose color from colorbar"

sns.set_context("poster",font_scale=0.6) 
sns.barplot(x="day",y="total_bill",data=var,palette="cool")
plt.show()
```

# Facet-Grid plot

```python
# It is use to represent different cateogries of paricular column with multiple plots according categories
# variable = sns.FacetGrid(var. of DS,col="column name",hue=" ")
# variable.map(plt.plot you want,"column name of x axis","column name of y axis").add_legend()
```

```python
sns.set_context("paper",font_scale=1) 
x = sns.FacetGrid(var,col="day",hue="sex")
x.map(plt.scatter,"total_bill","tip").add_legend()
plt.show()
```

```python
# To change the color we use palette="choose from color bar"
# To create a outline in plots we use edgecolor="color shortcut"

x = sns.FacetGrid(var,col="day",hue="sex",palette="cool")
x.map(plt.bar,"total_bill","tip",edgecolor="r").add_legend()
plt.show()
```
