```python
# How to give labels inside the bar plot
# The work of label is to indicate which color represent which data

plt.bar(x,y,label="ws")
plt.legend() # We apply plt.legend() with label function otherwise label function not works
plt.show()
```

```python
# How to outline bars with different color
# edgecolor is use to make outline and in input you write the color name shortcut like 'r','y','g','b',etc...

plt.bar(x,y,edgecolor='r',linewidth=2) # linewidth use to make outline more thicker
plt.show()
```

```python
# How to combine bar plots with multiple values
# You show that this bar plot is overlaping

z = [10,50,30,40,85]

plt.bar(x,y,color="r",width=0.2)
plt.bar(x,z,color="b",width=0.4)
plt.show() 
```

```python
# to avoid overlaping we follow the below programm
# you notice that the names of x axis is gone 
# to recover x axis names we follow below programm

import numpy as np
 
# This program is for seprate overlap data

width=0.2
a = np.arange(len(x))
a1 = [b+width for b in a]

#This programm is for join mutliple data into bar plot
plt.bar(a,y,width=0.2,label="one")
plt.bar(a1,z,width=0.2,label='two')

plt.xticks(a+width/2,x,rotation=5) # This program is to recover x axis names and rotate the names
plt.legend() # This programm is apply with label function
plt.show()
```

```python
# How to make horizontal bar plot

plt.barh(x,y)
plt.show()
```

```python
# How to change the transparency of bar plot
# By default the alpha is 1

plt.bar(x,y,alpha=0.2) # The value of alpha is between 0-1
plt.show()
```

# Linear Plot


```python
# How to make Linear Plot
# Example of Linear Plot

x = [1,2,5,4]
y = [8,10,4,2]

plt.plot(x,y)
plt.show
```

```python
# We import and use numpy here just to get some values for making linear plot. 

import numpy as np

x = np.arange(10,20)
y = x*x
```

```python
# How to make Linear plot 
# How to customize the color of line
# Here are some options available

# character        color
# =============    ===============================
# ``'b'``          blue
# ``'g'``          green
# ``'r'``          red
# ``'c'``          cyan
# ``'m'``          magenta
# ``'y'``          yellow
# ``'k'``          black
# ``'w'``          white

plt.plot(x,y,'m') 
plt.show()
```

```python
# **Line Styles**

# =============    ===============================
# character        description
# =============    ===============================
# ``'-'``          solid line style
# ``'--'``         dashed line style
# ``'-.'``         dash-dot line style
# ``':'``          dotted line style
# =============    ===============================
```

```python
# How to customize the line in the linear plot
# How to give the name to x, y axis and give title
# linewidth = use for customize the width of line 
# marker = use for customize the size of dots

plt.plot(x,y,'o:',linewidth=2,markersize=12)
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Linear Plot')
plt.show()
```

# Pie Plot

```python
x = [1250,1500,2230,850]
y = ['sameer','shuaib','aquib','faiz']

# labels parameter use to indicates the names. Which data of which person 
# labeldistance is used to set the distace of labels in pie plot
# By default your pie plot make counter clockwise diagram.To change into clockwise direction we use counterclock=False parameter
# by using counterclock parameter the diagram will show in clockwise and in ascending form

plt.pie(x,labels=y,labeldistance=1.2,counterclock=False)  
plt.show()
```

```python
# How to customize pie plot
# How to cut data of particular column. explode parameter is used for doing that
# How to customize colors. colors parameter is used for doing that
# How to get percentage of all inside pie plot. autopct='%0.2f%%' is used for doing that
# How to give shawdow to pie plot. Its looks better. shadow = True parameter is use for doing that
# How to customize the size of pie plot. radius parameter is used for doing that.
# How to change the starting point of pie plot. By default it starts from 0 degree. By startangle parameter we can do this.
# How to change the font size to look better. textprops={"fontsize":__} this parametere is use for doing that

ex = [0.2,0.0,0.0,0.0]

c = ['r','b','c','m','g']

plt.pie(x,labels=y,explode=ex,colors=c,autopct='%0.2f%%',shadow=True,radius=0.8,startangle=45
       ,textprops={"fontsize":14})

plt.show()                
```

```python
# How to make shadow more thicker. we use wedgeprops={"linewidth":___}
# How to change shadow color. We use wedgeprops={"edgecolor":"__"}
# How to rotate labels. We use rotatelabels=True
# How to give title to Pie Plot
# How to give labels/indicators in pie plot to understand which color indicates data of what. we use plt.legend() parameter

plt.pie(x,shadow=True,wedgeprops={"linewidth":5,"edgecolor":'m'},labels=y,rotatelabels=True)
plt.title("Example of Pie Plot")
plt.legend(loc=1) # loc = 1-10, we put in loc to change the location of labels inside pie plot
plt.show()
```

```python
# How to overlap pie plots to compare between them

x1 = [1000,1450,1680,1970]

c=['m','w','k','g','y']

plt.pie(x,labels=y,radius=1)
plt.pie(x1,radius=0.5,labels=y,colors=c)
plt.show()
```

# Stem Plot :- It is a combination of Linear Plot and Scatter Plot.

```python
# How to stem plot and give name to x, y axis and title

x = [5,7,8,3,1]
y = [7,5,1,6,3]

plt.stem(x,y)
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Stem Plot')
plt.show()
```

```python
# How to change the shape of lines in stem plot. We use linefmt='--',':', etc...
# How to change dots in stem plot. We use markerfmt='*','+', etc. with color shortcut
# Bottom parameter use for chang the starting point
# How to change the color of base line. By default it is red. We use basefmt='color shortcut'
# How to add label inside the stem plot. we use label parameter along with plt.legend() function otherwise label will not shown
# How to change the color of lines in line plot. We use use_line_collection=False because by default it is True

plt.stem(x,y,linefmt='--',markerfmt='m*',bottom=3,basefmt='y',label='hello world',use_line_collection=False)
plt.legend()
plt.show()
```

```python
# How to make horizontal stem plot. We use orientation="horizontal"

plt.stem(x,y,orientation="horizontal")
plt.show()
```

# Box Plot

```python
# How to make a boxplot

x = [10,20,30,40,50,60,70]

plt.boxplot(x)
plt.show()
```

```python
# How to make horizontal box plot

plt.boxplot(x,vert=False)
plt.show()
```

```python
# How to change the shape of box plot

plt.boxplot(x,notch=True)
plt.show()
```

```python
# How to change the size/width of box inside box plot. By default the width size is 0.2
# How to mark mean of data inside box plot. We use showmeans=True parameter
# How to change the color of box inside the box plot.We use boxprops={"color":"color shortcut"}. 
# By default the color of box is black
# How to change the color of upper and lower line of box called as caps. We use capprops={"color":"color shortcut"}

plt.boxplot(x,widths=0.3,showmeans=True,boxprops={"color":"m"},capprops={"color":"r"})

plt.show()
```

```python
# How to change the name of label. By default it is 1. We use labels=["___"]
# How to fill color inside the box. We use patch_artist=True
# How to change the color of line which join caps with box called as whisker line. We use whiskerprops={"color":"_"} 
# By default the color of whisker line is black 

plt.boxplot(x,labels=['Hello'],patch_artist=True,whiskerprops={"color":"m"})
plt.show()
```

```python
# You notice that their is a dot on the top which indicates that the data is not in organised manner.
# Some data is exceeding the range. This 'DOT' is called "OUTLIER"

x = [5,10,15,20,25,50]

plt.boxplot(x)
plt.show()
```

```python
# How to change the color and shape of outlier

plt.boxplot(x,sym='m*')
plt.show()
```

```python
# To connect the data of outlier with line we use whis=___. 
# The value given with whis parameter is for joing the line with outlier. you can give value according to data like 1,2,3,4,5...
# showmeans=True parameter mark a symbol at the place of mean of data

plt.boxplot(x,whis=5,showmeans=True)
plt.show
```

```python
# How to plot multiple boxes of multiple data inside the box plot and change the label name. By Default it is 1,2,3,4...

x = [10,20,30,40,50]
x1 = [20,40,60,80,100]
x2 = [10,30,60,90,120,150]

y = [x,x1,x2]

plt.boxplot(y,labels=["C","C++","python"]) # while using labels parameter given new label names are equivalent to label values
plt.show()                                 
```

# Stack Plot or Area Plot

```python
# How to make a Stack Plot

x = [1,2,3,4,5]
area = [10,30,50,25,20]

plt.stackplot(x,area)
plt.xlabel('X Axis')
plt.ylabel('Area')
plt.title('Stack Plot')
plt.show()
```

```python
# How to make multiple Stack plots of multiple data inside a single Stack plot.
# How to make labels for each data to understand what color indicates which data. 
# Always use plt.legend() function with labels parameter otherwise labels parameter not work.
# How to change the colors of data. We use colors=['_','_','_'] parameter

x = [15,20,25,30,35]
area1 = [10,54,79,42,5.5]
area2 = [5,86,54,15.25,20]
area3 = [ 15,24,73,98,23]

l = ['area1','area2','area3']

plt.stackplot(x,area1,area2,area3,labels=l,colors=['c','m','k'])
plt.legend(loc=2) # To change the location of labels we use loc=1-10 inside legend function
plt.show()
```

```python
# How to change the base line of stack plot. We use baseline="__" parameter 
# Their are 2 values which we give inside the baseline parameter to change baseline :- "sym","wiggle"
# We can make grids inside the stack plot with the help of plt.grid() function


plt.stackplot(x,area1,area2,area3,baseline="sym")
plt.grid()
plt.show()
```

# Step Plot

### All the parameters of Linear Plot works same in Step Plot 

```python
# How to make step plot and give name to x, y axis and title

x = [1,2,3,4,5]
y = [55,44,78,3,1]

plt.step(x,y)
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Step Plot')
plt.show()
```

```python
# How to make grid inside a step plot to understand data better. We use plt.grid() function
# How to change the line inside the step plot. We use color=" color shortcut"
# How to mark the dots inside the step plot to understand data better. We use marker='o','*', etc...
# We can change the size of dots. We use ms=__
# We can change the color of dots. We use mfc='color shortcut'. By default the color of dot is same as line color
# How to make label for understand what color indicates which data. We use label="__". 
# We use plt.legend() function with label parameter otherwise label will not shown

plt.step(x,y,color='m',marker="o",ms=20,mfc='r',label="Data Science")
plt.grid()
plt.legend(loc=2) # We use loc=1-10 to change the location of label. By default label shown at the top right
plt.show()
```

# Fill_Between Plot

# Fill_between plot is used to fill color inside the plots

## We can use fill_between plot with any plot but usually we can use this with Linear Plot

```python
# Make a LInear Plot to understand this topic better
# We import numpy because some parameters of fill_between plot doesn't work with list data type

import numpy as np

x = np.array([1,2,3,4,5])
y = np.array([1,2,3,4,5])

plt.plot(x,y)
plt.show()
```

```python
# How to use fill_between plot

plt.plot(x,y)
plt.fill_between(x,y)
plt.show()
```

```python
# How to set parameters in fill_between plot 

plt.plot(x,y)
plt.fill_between(x=[2,4],y1=2,y2=4) #from where to where fill_between plot works inside. We fill values according x, y1, y2 axis
plt.show()
```

```python
# We use where=(x..) & (x..) parameter for doing the same task
# The advantage of using where parameter is you don't need to give the value of y axis. You just give 2 parameter of x axis only

plt.fill_between(x,y,where=(x>2) & (x<5))
plt.show()
```

```python
# How to change color of fill_between plot. We use color="color shortcut"
# How to change the brightness inside the plot. We use alpha=__ parameter to adjust the brightness. By default it is 1.

plt.fill_between(x,y,color='r',alpha=0.5)
plt.show()
```

# Subplot 

```python
 Parameters
----------
*args : int, (int, int, *index*), or `.SubplotSpec`, default: (1, 1, 1)
    The position of the subplot described by one of

    - Three integers (*nrows*, *ncols*, *index*). The subplot will take the
      *index* position on a grid with *nrows* rows and *ncols* columns.
      *index* starts at 1 in the upper left corner and increases to the
      right. *index* can also be a two-tuple specifying the (*first*,
      *last*) indices (1-based, and including *last*) of the subplot, e.g.,
      ``fig.add_subplot(3, 1, (1, 2))`` makes a subplot that spans the
      upper 2/3 of the figure.
    - A 3-digit integer. The digits are interpreted as if given separately
      as three single-digit integers, i.e. ``fig.add_subplot(235)`` is the
      same as ``fig.add_subplot(2, 3, 5)``. Note that this can only be used
      if there are no more than 9 subplots.
    - A `.SubplotSpec`.
```

```python
x = [10,20,30,40]
y = [520,80,20,45]

plt.plot(x,y)

plt.pie(x)

plt.scatter(x,y)

plt.bar(x,y)

plt.show()
```

```python
# Notice that all the plot overlaping
# To avoid overlaping of different plots of same data we use subplot(nrows,ncols,index) .By default it is (1,1,1)
# By using subplot different plots will generate at same place of same data as well as of different data

plt.subplot(2,2,1)
plt.stackplot(x,y)

plt.subplot(2,2,2)
plt.pie(x)

plt.subplot(2,2,3)
plt.scatter(x,y)

plt.subplot(2,2,4)
plt.bar(x,y)

plt.show()
```

# Text in Matplotlib

# text :- Add text in an arbitary location of the Axis.

# annotate :- Add an annotation, with an optional arrow, at an arbitary location                         the Axis

# xlabel :- Add a label to x axis

# ylabel :- Add a label to y axis

# title :- Add a title to the axis

```python
# How to give name to the x, y axis and title and change their font size. The default font size is 10
# How to insert a text inside the plot to understand the data better. 
# We use plt.text(value of x axis, value of y axis, "________")
# How to change the size of insertes text. We use fontsize=__. The default fontsize is 10.
# We can change the font style by using style="font style"

plt.plot(x,y)
plt.xlabel("X Axis",fontsize=20)
plt.ylabel("Y Axis",fontsize=20)
plt.title("Title",fontsize=20)
plt.text(2,4,"Data Science",fontsize=15,style="italic")
plt.show()
```

```python
# How to create a box outside the inserted text and color them
# We use bbox={"facecolor":"full color name"}

plt.plot(x,y)
plt.xlabel("X Axis",fontsize=20)
plt.ylabel("Y Axis",fontsize=20)
plt.title("Title",fontsize=20)
plt.text(2,4,"Data Science",fontsize=15,bbox={"facecolor":"red"})
plt.show()
```

```python
# How to insert a text inside and create a arrow for them
# We use plt.annotate("Text",xy(where the arrow points according x,y axis),xytext(postion of text according x,y axis)
# How to customize the color of arrow. We use arrowprops={"facecolor":"color shortcut"}
# How to insert a label inside the plot to representthe data and customize their position inside the plot
# We use plt.legend(["label name",loc=0-10]). loc is used to change the position of label

plt.plot(x,y)
plt.annotate("Ekeeda",xy=(3,2),xytext=(4,4),arrowprops={"facecolor":"g","shrink":20})
plt.legend(["hello"],loc=1)
plt.show()
```

```python
# How to fill color inside the label box. We use plt.legend("labeltext", facecolor="color shortcut")
# How to change edgecolor of label box. By Default the edgecolor of label box is same as shadow color
# We use plt.legend("labeltext", edgecolor="color shortcut")
# How to change the brightness of label box. By default the brightness is 1. We use frameaplha= 0-1 
# How to give shadow to label box. We use plt.legend("labeltext", shadow = True)

plt.plot(x,y)
plt.annotate("Ekeeda",xy=(3,2),xytext=(4,4),arrowprops={"facecolor":"g","shrink":20})
plt.legend(["hello"],loc=1,facecolor="m",edgecolor="y",framealpha=0.5, shadow = True)
plt.show()
```
