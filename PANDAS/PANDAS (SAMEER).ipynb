# Types of data structures in Pandas

# 1) Series :- It is defined as one dimensional array that is capable of storing various data types. We can create series with any python data type like set, tuple, dictionary, list.


# pd.Series() 

# Index :- The vertical values of first column is known as index.

```python
# Example of making a series with list data type.
x = [1,5,7,5,2,7,3,89]
a = pd.Series(x)
print(a)
print()
print(type(a)) # to check data type we use type()
print()
print(a[5]) # to check the item in series with the help of their index value
```

```python
# how to change the index of a series
x = [1,5,7,5,9]
a = pd.Series(x,index=["a","b","c","d","e"])
print(a)
```

```python
# how to change data type
a = [7,5,2,4]
x = pd.Series(a,dtype=float)
print(x)
```

```python
#how to name a data type 
a = [4,5,2,4]
x = pd.Series(a,name="sameer")
print(x)
```

```python
# Example of making a series with dictionary
x = {"name":["sameer","faiz","shuaib"],"age":[22,23,20],"location":["nepal","sitapur","sitapur"]}
a = pd.Series(x)
print(a)
print()
print(type(a))
```

```python
# Example of making series with single value
x = pd.Series(12)
print(x)
print(type(x))
```

```python
# Example of making single value series with multiple indexes.
x = pd.Series(12,index=[1,2,3,4,5,6,7,8,9])
print(x)
print(type(x))
```

```python
# Example of addition of 2 series.
# When the values are missing it show NaN.
# When we perform same task in numpy it shows broadcasting error. It's the difference between pandas and numpy.
# We can add different indexes series with one another but it is not possible in numpy.
# In numpy we can add only same indexes arrays.

x = pd.Series(12,index=[1,2,3,4,5,6,7,8,9])

x1 = pd.Series(12,index=[1,2,3,4,5])

print(x+x1)
```

# 2) Data Frame / pd.DataFrame() :- we can mainly create data frame with dictionary and list data type but we can create it with any data type.

# Merge :- Merge function use to merge data sets with minimum one similar column name. It only merge Data Sets Horizonatlly not Vertically.

```python
# Remember that one similar column is available to join another data set on 

x = pd.DataFrame({"Name":["sharukh","salman","amir","john","amitabh","govinda"],"age":[25,85,45,35,75,25]})

y = pd.DataFrame({"Name":["sharukh","salman","amir","john","amitabh","govinda"],"shade_no.":[12,10,75,12,42,23]})

pd.merge(x,y,on="Name") #pd.merge(var1,var2,on="similar column name")
```

```python
# If you want that the data of second data set shown first

pd.merge(y,x)
```

```python
# It only merges the similar values in both the data set

x = pd.DataFrame({"Name":["sharukh","salman","amir","john","amitabh","govinda"],"age":[25,85,45,35,75,25]})

y = pd.DataFrame({"Name":["sharukh","salman","amir","john","amitabh","johny lever"],"shade_no.":[12,10,75,12,42,23]})

pd.merge(x,y)
```

```python
# It show all the data from right Data Set merges with similar data of left data set

pd.merge(x,y,how="right")
```

```python
# It shows all the data of left Data Set and similar data of right Data Set

pd.merge(x,y,how="left")
```

```python
# It show all data of both the Data Set Whether it is similar or not

pd.merge(x,y,how="outer")
```

```python
# It shows that the which data is merging from which Data Set

pd.merge(x,y,how="outer",indicator=True)
```

```python
# If the column names of both Data Set are same. So, it will only show the column names and the data is not merged.

x = pd.DataFrame({"Name":["sharukh","salman","amir","john","amitabh","govinda"],"age":[25,85,45,35,75,25]})

y = pd.DataFrame({"Name":["s","s","a","j","a","g"],"age":[12,10,75,12,42,23]})

pd.merge(x,y)
```

```python
# If you want to merge the data of both Data sets where column names are same whether it contains same data or different.

pd.merge(x,y,left_index=True,right_index=True)
```

```python
# If you want to rename the columns of merged Data sets you can do.
# The new given column names are merged with old data sets column names.

pd.merge(x,y,left_index=True,right_index=True,suffixes=("SAM","DEF"))
```

# Join 

```python
# Lets crete 2 Data Sets for using .join() function

x = pd.DataFrame({"Name":["sameer","faiz","abdullah","shuaib","aquib"],"marks in 10th":[98,95,87,89,76]})
y = pd.DataFrame({"city":["dhangadhi","sitapur","lucknow","sitapur","bahraich"],"marks in 12th":[78,87,97,79,85]})

x.join(y)
```

```python
# When we work in 2 different Data sets where data is not equivalent
# Join function works

x = pd.DataFrame({"Name":["sameer","faiz","abdullah","shuaib","aquib"],"marks in 10th":[98,95,87,58,78]})
y = pd.DataFrame({"city":["dhangadhi","sitapur","lucknow"],"marks in 12th":[78,87,85]})

x.join(y)
```

```python
# How to change the index value while joining 2 data sets
# index of the both data sets must be same other wise it will not work
# index is equivalent in lenth to old index otherwise it will show error

x = pd.DataFrame({"Name":["sameer","faiz","abdullah","shuaib","aquib"],"marks in 10th":[98,95,87,89,76]},index=["a","b","c","d","e"])
y = pd.DataFrame({"city":["dhangadhi","sitapur","lucknow"],"marks in 12th":[78,87,97]},index=["a","b","c"])

x.join(y)
```

```python
# If we join the Data Sets in vice-verse ìt will join

y.join(x)
```

```python
# it will join the data sets from left side
# left join works same as join by default
y.join(x,how = "left")
```

```python
# It will join the data sets from right side.

y.join(x,how = "right")
```

```python
# It will join the Data Sets from outer side

y.join(x,how = "outer")
```

```python
# It will join the data sets from inner side
# Inner works same as join by default

y.join(x,how="inner")
```

```python
# If any column name is similar in both the Data Sets it will show error while joining. 

x = pd.DataFrame({"Name":["sameer","faiz","abdullah","shuaib","aquib"],"marks in 10th":[98,95,87,89,76]})
y = pd.DataFrame({"Name":["dhangadhi","sitapur","lucknow","sitapur","bahraich"],"marks in 12th":[78,87,97,79,85]})

x.join(y)
```

```python
# To remove this error we use suffix function
# Their are 2 suffix functions 
# rsuffix :- it will change the similar column name of right data set at the time of joining.
# lsuffix :- it will change the similar column name of left data set at the time of joining.
# The new given name is get combine with the old one.

x.join(y,rsuffix="_1")
```

```python
x.join(y,lsuffix="_2")
```

```python
x.join(y,rsuffix="_1",lsuffix="_2")
```

```python
# In join function if the name of columns are similar in Data Sets it will show error but in append this will not happen

x = pd.DataFrame({"Name":["sameer","faiz","abdullah","shuaib","aquib"],"marks in 10th":[98,95,87,89,76]})
y = pd.DataFrame({"Name":["dhangadhi","sitapur","lucknow","sitapur","bahraich"],"marks in 12th":[78,87,97,79,85]})

x.append(y)
```

```python
# If the data is unorganised it will join data sets

x = pd.DataFrame({"Name":["sameer","faiz","abdullah","shuaib","aquib"],"marks in 10th":[98,95,87,89,76]})
y = pd.DataFrame({"City":["dhangadhi","sitapur","lucknow"],"marks in 12th":[78,87,97]})

x.append(y)
```

```python
# You have noticed that index numbers are overlaping due to unorganised data 
# You have option to organised the index numbers

x.append(y,ignore_index = True)
```

# Melt() :- It is use to reshape the Data Sets

```python
x = pd.DataFrame({"name":["sameer","faiz","shuaib","aquib"],"city":["dhangadhi","sitapur","sitapur","bahraich"],"code":[45,78,98,15]})

x
```

```python
# How to reshape Data Set vertically
# By default all the data is converted into 2 columns.
# All the values show in value column and all the column names are show in variable column.

pd.melt(x)
```

```python
# How to make any column as a index in vertical Data Set

pd.melt(x,id_vars=["code"])
```

```python
# How to change the name of variable column

pd.melt(x,var_name="hello")
```

```python
# IHow to change the name of value column 

pd.melt(x,value_name="mannu")
```

# Pivot()

```python
x = pd.DataFrame({"name":["sameer","faiz","shuaib","sameer","faiz","shuaib"],
                  "eng":[45,15,58,75,25,65],"hindi":[45,78,98,15,48,95],"roll no":[4,6,8,9,10,7]})

x
```

```python
# When you use pivot function directly it will show error
# To avoid error you have to pass arguments

x.pivot()
```

```python
# you have to pass two arguments
# index and columns

x.pivot(index="roll no",columns="name")
```

```python
# How to get data of particular column

x.pivot(index="roll no",columns="name",values="eng")
```

```python
# how to calculate mean of the given Data set

x = pd.DataFrame({"name":["sameer","faiz","sameer","sameer","faiz","faiz"],
                  "eng":[45,15,58,75,25,65],"hindi":[45,78,98,15,48,95],"roll no":[4,6,8,9,10,7]})

x.pivot_table(index="name",columns="roll no",aggfunc="mean")
```

```python
# How to calculate sum

x.pivot_table(index="name",columns="roll no",aggfunc="sum")
```

```python
# How to calculate the margin (mean) of Data set

x.pivot_table(index="name",columns="roll no",margins="True")
```
