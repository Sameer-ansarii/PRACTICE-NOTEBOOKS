# Dropna

## .dropna() :- it drops or removes row/column where any value is missing.

```python
x.dropna()
```

```python
# It drops all rows where any value is missing

x.dropna(how="any")
```

```python
# It drops rows where all values are "NaN"

x.dropna(how="all")
```

```python
# How to remove 'NaN' values of a particular column's rows

# we use .dropna(subset=["column name"])

# it removes rows where 'NaN' is in given column name

x.dropna(subset=["listedproducts"])
```

# fillna() :-  it is use to fill values at place of null values in Data Set

```python
# How to fill value at 'NaN' Place

x.fillna("panda")
```

```python
# How to fill different values in different columns

x.fillna({"totalurgencycount":"sameer","urgencytextrate":"ansari"})
```

```python
# There are two methods of filling the 'NaN' values
# Fordward filling or "ffill":- It fills the value with the previous value
# Backward or "bfill":- It fills the value with the next value
# we can apply this on the basis of axis

x.fillna(method="ffill")
```

```python
x.fillna(method="bfill")
```

```python
x.fillna(method="ffill",axis=0)
```

# Merge

## Merge function use to merge data sets with minimum one similar column name. It only merge Data Sets Horizonatlly not Vertically.

```python
# Add 2 data sets for practice

x = pd.read_csv("C:\\Users\\HP\\Downloads\\aa.csv")
y= pd.read_csv("C:\\Users\\HP\\Downloads\\bb.csv")
```

```python
y.head(2)
```

```python
# Remember that one similar column is must available to join another data set on 

pd.merge(x,y,on="totalunitssold")
```

```python
# It will only merge data sets with equivalent length 

pd.merge(y,x)
```

# # Concat :- It merges Data Sets whether their column names are same or not. It also use to merge Data sets horizonatlly or vertically, base on thier axis.

```python
# It merges Data Sets whether their column names are similar or not.
# How to merge 2 Data Sets vertically

pd.concat([x,y])
```

```python
# How to merge 2 Data Sets Horizontally

pd.concat([x,y],axis=1)
```

```python
# It merge Data Sets and automatically removes columns where any value is 'NaN'

pd.concat([x,y],join="inner")
```

```python
# It merge Data Sets and automatically removes rows where any value is 'NaN'

pd.concat([x,y],join="outer")
```

```python
# How to the create the key / name of data sets
# We can also use this function on the basis of their axis

pd.concat([x,y],keys=["sameer","ansari"])
```

# Join

```python
# Lets import 2 Data Sets for practice
# It will not join Data Sets where any column name is similar

x = pd.read_csv("C:\\Users\\HP\\Downloads\\aa.csv")
y = pd.read_csv("C:\\Users\\HP\\Downloads\\unique-categories.sorted-by-count.csv")

x.join(y)
```

```python
# Inner works same as join by default

x.join(y,how="inner")
```

```python
x.join(y,how="outer")
```

```python
# left join works same as join by default

x.join (y,how="left")
```

```python
x.join(y,how="right")
```

# Append :- It will join two Data Sets and make a new Data set.

```python
# In join function if the name of columns are similar in Data Sets it will show error but in append this will not happen

x.append(y)
```

```python
# You have noticed that index numbers are overlaping due to unorganised data 
# You have option to organised the index numbers

x.append(y,ignore_index=True)
```

# Melt() :- It is use to reshape the Data Sets

```python
# How to reshape Data Set vertically
# By default all the data is converted into 2 columns.
# variable = column names, value = rows
# All the values show in value column and all the column names are show in variable column.

pd.melt(x)
```

```python
# How to make any column as a index in vertical Data Set

pd.melt(x,id_vars=["index"])
```

```python
# How to change variable and column name

pd.melt(x,var_name="sameer",value_name="ansari")
```
